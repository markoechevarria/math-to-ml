1. 3D Broadcasting for a "Splat" Effect: Create a 3D NumPy array representing a grid of coordinates, for example, a shape of (100, 100, 3). Now, define a single point (a 1D array of shape (3,)) representing a light source. Using 3D broadcasting, calculate the Euclidean distance from every point in the grid to the light source. This will create a "splat" or gradient effect that can be used for things like procedural texture generation.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
%matplotlib widget
```

```python
# creating the 3d grid
x = np.linspace(0, 200, 10)
y = np.linspace(0, 200, 10)
z = np.linspace(0, 200, 10)
X,Y,Z = np.meshgrid(x, y, z)
points_array = np.vstack([X.flatten(), Y.flatten(), Z.flatten()]).T

# creating the point "light source"
point = np.array([200,200,200])
point = ( np.random.random(3) * 199 ) + 1

# first way to compute euclidean distance
distance = np.linalg.norm(points_array - point, axis=1)

# second way to compute euclidean distance
distancia_aux = (points_array - point) ** 2
distance = np.sqrt( distancia_aux[:,0] + distancia_aux[:,1] + distancia_aux[:,2] )

# plt configuration
plt.close('all')
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

# color scale
scatter = ax.scatter(points_array[:, 0], points_array[:, 1], points_array[:, 2], c=distance, s=50)

# plot the point
ax.scatter(point[0], point[1], point[2], color="red")

# Color bar
fig.colorbar(scatter, ax=ax, label='Euclidean distance')

# legend
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Splat Effect')
plt.show()
```

# Boolean indexing

Calculate the prime numbers between 0 and 100 by using a Boolean array.

```python
is_prime = np.ones((100,), dtype=bool)
is_prime[:2] = 0

nmax = int(np.sqrt(len(is_prime)))
for i in range(2, nmax):
    print(is_prime)
    is_prime[2*i::i] = False

print(np.nonzero(is_prime))
```
